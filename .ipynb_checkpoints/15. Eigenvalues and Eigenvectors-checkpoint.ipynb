```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("seaborn-poster")

%matplotlib inline

def plot_vect(x, b, xlim, ylim):
    plt.figure(figsize = (10,6))
    plt.quiver(0,0,x[0],x[1],
          color="k",angles="xy",
          scale_units="xy",scale=1,
          label="Original vector")
    plt.quiver(0,0,b[0],b[1],
          color="g",angles="xy",
          scale_units="xy",scale=1,
          label="Transformed vector")
    plt.xlim(xlim)
    plt.ylim(ylim)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.legend()
    plt.show()
```

```python
A = np.array([[2,0],[0,1]])

x = np.array([[1],[0]])
b = np.dot(A, x)
plot_vect(x,b,(0,3),(-1,1))
```

```python
import numpy as np
from numpy.linalg import inv

def normalize(x):
    fac = abs(x).max()
    x_n = x / x.max()
    return fac, x_n

x = np.random.rand(2)
y = np.array([0.1,1])
a = np.array([[0,2],[2,3]])

for i in range(10):
    x = np.dot(a, x)
    lambda_1, x = normalize(x)
  

print("Eigenvalue:", lambda_1)
print("Eigenvector:", x, "\n")

for i in range(10):
    y = np.dot(inv(a), y)
    lambda_2, y = normalize(y)
    
print("Eigenvalue:", lambda_2)
print("Eigenvector:", y, "\n")   
```

```python
import numpy as np
from numpy.linalg import qr

a = np.array([[0,2],
            [2,3]])

p = [1, 5, 10, 20]

for i in range(20):
    q, r = qr(a)
    a = np.dot(r, q)
    if i+1 in p:
        print(f"Iteration #{i+1}:")
        print(a)
```

```python
import numpy as np
from numpy.linalg import eig

a = np.array([[0,2],
             [2,3]])

w,v = eig(a)
print(w)
print(v)
```
